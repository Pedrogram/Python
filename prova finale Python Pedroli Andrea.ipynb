2. Si chiede di trovare, per ogni continente:
 a. il numero di casi fin dall'inizio della pandemia
 b. la percentuale rispetto al totale mondiale del numero di casi

```python
#a. Dopo aver appurato che il valore Total Cases è cumulativo di tutti i new cases, calcolo il valore massimo dei casi totali per ogni stato
totali_per_stato = dataset.groupby("location")["total_cases"].max().reset_index()

#accoppiamo continente e stato, eliminando duplicati e usando una left join, anche se non ci sono Null
dataset_unique = dataset[['location', 'continent']].drop_duplicates()
totali_per_stato = totali_per_stato.merge(dataset_unique, on="location", how="left")

# Raggruppamento per continente e somma degli stati
totali_per_continente = totali_per_stato.groupby("continent")["total_cases"].sum()

# Visualizzazione dei risultati
print(totali_per_continente)
```

```python
# b. Arrotondo senza decimali le percentuali
totale_mondiale = totali_per_continente.sum()
percentuali_continente = (totali_per_continente / totale_mondiale * 100).round(0)

print("\nPercentuali dei Casi Totali per Continente:\n", percentuali_continente)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filtro per Italia, Germania e Francia dal maggio 2022 all'aprile 2023
countries = ['Italy', 'Germany', 'France']
filtered_data = dataset[(dataset['location'].isin(countries)) &
                        (dataset['date'] >= '2022-05-01') & 
                        (dataset['date'] <= '2023-04-30')]

# Rimozione delle righe senza dati ICU
filtered_data = filtered_data.dropna(subset=['icu_patients'])

# Creazione del boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='location', y='icu_patients', data=filtered_data)
plt.title('Pazienti in ICU per Italia, Germania e Francia (Maggio 2022 - Aprile 2023)')
plt.xlabel('Stato')
plt.ylabel('Numero Pazienti in ICU')
plt.show()
```

COMMENTO: Nell'intervallo considerato, Germania e Francia hanno registrato un numero di pazienti in terapia intensiva superiore rispetto all'Italia. I valori estremi osservati nei boxplot per Germania e Francia indicano la presenza di picchi significativi nella gestione dei casi di COVID-19, che sono stati molto meno accentuati in Italia.

```python
# Filtraggio per le nazioni e per l'anno 2023
selected_countries = ['Italy', 'Germany', 'France', 'Spain']
dataset["date"] = pd.to_datetime(dataset["date"])
hospital_data_2023 = dataset[(dataset['location'].isin(selected_countries)) & (dataset['date'].dt.year == 2023)]

# Calcolo della somma dei pazienti ospitalizzati per ciascuna nazione
total_hospitalized = hospital_data_2023.groupby('location')['hosp_patients'].sum()

# Verifica dei dati nulli
missing_data = hospital_data_2023['hosp_patients'].isna().sum()

# Mostrare i risultati
print("Somma dei Pazienti Ospitalizzati per Nazione nel 2023:\n", total_hospitalized)
print("\nDati Mancanti per Pazienti Ospitalizzati nel 2023: ", missing_data)
```

COMMENTO: Per la Germania non ci sono dati per l'intero 2023, si può dedurre che sia una mancata rilevazione e non che non ci fossero Pazienti Ospedalizzati, si puà pensare di stimare gli ospedalizzati a partire dalle terapie intensive, vedendo negli altri paesi come variano in percentuale questi 2 dati. Come da calcoli che seguono, possiamo prendere la Francia come paragone più adatto alla Germania e ipotizzare che il numero di Terapie intensive del 2023 sia il 6% del totale degli ospitalizzati.

```python
import pandas as pd

# Assicuriamoci che la colonna 'date' sia in formato datetime
dataset['date'] = pd.to_datetime(dataset['date'])

# Filtraggio per Francia e Italia per gli anni 2022 e 2023
countries = ['France']
years = [2022, 2023]
filtered_data = dataset[(dataset['location'].isin(countries)) & (dataset['date'].dt.year.isin(years))]

# Assicurarsi che non ci siano valori zero che potrebbero causare divisione per zero
filtered_data = filtered_data[(filtered_data['icu_patients'] > 0) & (filtered_data['hosp_patients'] > 0)]

# Calcolo del rapporto in percentuale tra pazienti in ICU e ospedalizzati
filtered_data['percentage_ratio'] = (filtered_data['icu_patients'] / filtered_data['hosp_patients']) * 100

# Aggregazione dei dati per paese e anno
results = filtered_data.groupby([filtered_data['date'].dt.year, 'location'])['percentage_ratio'].mean().unstack()

# Visualizzazione dei risultati in forma tabellare
print("Rapporto Medio Annuo in Percentuale tra Pazienti in ICU e Ospedalizzati per Francia (2022-2023):")
print(results)


```
